```python
L=[22,25,25,25,26,26,27,29,30,31,31,31,102]

import matplotlib.pyplot as plt
plt.boxplot(L)
plt.show()
```

```python
import numpy as np

Q1=np.percentile(L,25)
Q3=np.percentile(L,75)
IQR=Q3-Q1
print(IQR)
print(Q1)
print(Q3)
```

```python
lower_outlier=Q1-(1.5*IQR)
upper_outlier=Q3+(1.5*IQR)
print(lower_outlier)
print(upper_outlier)
```

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/BostonHousing.csv')
```

```python
Q1=np.percentile((df["crim"]),25)
Q3=np.percentile((df["crim"]),75)

IQR=Q3-Q1
print(IQR)
print(Q1)
print(Q3)
```

```python
plot=plt.boxplot(df["crim"])
```

```python
count_lower=Q1-(1.5*IQR)
count_upper=Q3+(1.5*IQR)
print(count_lower)
print(count_upper)
```

```python
L = df["crim"]
```

```python
outlier = L[(L<count_lower) | (L>count_upper)]
```

```python
clean_data = L[(L>count_lower) | (L<count_upper)]
```

```python
len(clean_data) / len(outlier)
```
