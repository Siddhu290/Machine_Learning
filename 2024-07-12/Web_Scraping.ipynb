###Recursive Definations
Every recursive function includes two parts

*   Base case(s)(non-recursive)
*   Recursive Cases

```python
def factorial(n):
  if n==0:
    return 1
  else:
    return n*factorial(n-1)
```

##**Palindrom or not**




```python
def is_palindrome(n):
  if n < 0:
    return False
  if n < 10:
    return True
  s = str(n)
  if s[0] == s[-1]:
    return is_palindrome(int(s[1:-1]))
  else:
    return False
```

```python
is_palindrome(121)
```

```python
def sum(n):
  if n == 0:
    return 0
  else:
    return n + sum(n-1)
sum(5)
```

```python
def sort_l(a):
  for i in range(len(a)):
    for j in range(i + 1, len(a)):
      if a[i] > a[j]:
        a[i], a[j] = a[j], a[i]
  return a
a = [10, 2, 5, 30, 7]
sort_l([10, 2, 5, 30, 7])
```

```python
def gcd(a, b):
  if b == 0:
    return a
  else:
    return gcd(b, a % b)
gcd(144, 1024)
```

```python


def lcm(a, b):
  gcd_ab = gcd(a, b)


  lcm_ab = (a * b) // gcd_ab

  return lcm_ab

lcm(144, 1024)
```
