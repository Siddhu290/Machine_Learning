```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/BahramJannesar/ChocolateReveiwDataAnalysis/master/datasets/chocolate.csv')
df.head()
```

```python
df.head()
```

```python
new_name={'company_location':'location',
          'review_date' : 'date'}
df.rename(columns=new_name,inplace=True)
```

```python
df.head()
```

```python
taste1 = df[['first_taste', 'second_taste', 'third_taste', 'fourth_taste']]
taste1.head()
```

```python
taste1
```

```python
taste1.isna().sum()
taste1 = taste1.dropna()
taste1
```

```python
taste1.groupby('first_taste').count()
```

```python
taste1.groupby('first_taste').count()["second_taste"].sort_values(ascending=False)
```

```python
taste1.groupby('first_taste').count()["third_taste"].sort_values(ascending=False)
```
