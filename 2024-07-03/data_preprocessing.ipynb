```python
university_towns=[]
with open("/content/sample_data/university_towns.txt") as file:
  for line in file:
    if '[edit]' in line:
      state=line
    else:
      university_towns.append((state,line))
university_towns[:5]
```

```python
town_df=pd.DataFrame(university_towns,columns=['State','RegionName'])
```

```python
town_df
```

```python
town_df.head()
```

```python
town_df.describe()
```

```python
def get_city_state(item):
  if '(' in item:
    return item[:item.find('(')]
  elif '[' in item:
    return item[:item.find('[')]
  else:
    return item
```

```python
town_df = town_df.applymap(get_city_state)
```
