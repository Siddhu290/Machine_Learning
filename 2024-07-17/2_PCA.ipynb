```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv")
```

```python
df
```

```python
col=["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]
df.columns=col
```

```python
df
```

```python
from sklearn.decomposition import PCA
```

```python
X=df.iloc[:,0:13]
```

```python
pca=PCA(n_components=1)
```

```python
x1=pca.fit_transform(X)
```

```python
y=df.iloc[:,13].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y,random_state=0,test_size=0.2)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
