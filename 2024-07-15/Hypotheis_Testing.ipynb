```python
def main():
    while True:
        try:

            expression = input("Enter a Python expression to evaluate (or type 'exit' to quit): ")

            if expression.lower() == 'exit':
                print("Exiting the program.")
                break


            result = eval(expression)


            print(f"The result of '{expression}' is: {result}")

        except Exception as e:
            print(f"Error: {e}")

if __name__ == "__main__":
    main()

```

```python
def evaluate_expression(expression):

    expression = expression.replace(" ", "")


    def evaluate_sub_expression(sub_expr):
        stack = []
        for char in sub_expr:
            if char == ')':

                temp_expr = []
                while stack:
                    top = stack.pop()
                    if top == '(':
                        break
                    temp_expr.append(top)

                temp_expr.reverse()

                result = evaluate_expression(''.join(temp_expr))

                stack.append(str(result))
            else:
                stack.append(char)

        return eval(''.join(stack))


    if '(' not in expression and ')' not in expression:
        return eval(expression)

    return evaluate_sub_expression(expression)


expression = "3+(2*(5+9))+8"
result = evaluate_expression(expression)
print(f"The result of '{expression}' is: {result}")
```

##write a program to find all the characters which are repeated more than 5 times

```python
def find_repeated_characters(sentence):

    char_count = {}


    for char in sentence:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1


    repeated_characters = [char for char, count in char_count.items() if count > 5]

    return repeated_characters


sentence = "Hello, world! This is a sample sentence to demonstrate the program."
result = find_repeated_characters(sentence)
print("Characters repeated more than 5 times:", result)
```

## write a program to Check first number is any power of second number



```python
def is_power(base, num):
    if base == 1:
        return num == 1
    if base <= 0 or num <= 0:
        return False
    power = 1
    while power < num:
        power *= base
    return power == num


base = int(input("Enter the base number: "))
num = int(input("Enter the number to check: "))
if is_power(base, num):
    print(f"{num} is a power of {base}.")
else:
    print(f"{num} is not a power of {base}.")
```
