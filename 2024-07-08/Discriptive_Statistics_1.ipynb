### Variable Types

#### Nominal Variables
- Unordered categorical variables
- **Examples:** Survey responses: sex (M/F), true or false (T/F), yes or no (Y/N), color

#### Ordinal Variables
- There is an ordering but no implication of equal distance between the different points of the scale.
- **Examples:** Likert scale of 1 to 5 (comfort with programming), educational level (high school, some college, degree, graduate), size (S/M/L/XL)

#### Interval Variables
- There are equal differences between successive points on the scale but the position of zero is arbitrary.
- **Examples:** Temperature measurement in Celsius or Fahrenheit

#### Ratio Variables
- The relative magnitudes of scores and the differences between them matter. The position of zero is fixed.
- **Examples:** Absolute temperature (Kelvin scale), age, weight, length

## The Chi-Square Distribution ()

The chi-square distribution (), also known as the chi-squared distribution, is a fundamental tool in statistical inference. It describes the probability distribution of the sum of squared deviations from expected values, frequently encountered in hypothesis testing.

**Key Concepts:**

* **Degrees of Freedom (df):** This parameter influences the shape of the chi-square distribution. A higher number of degrees of freedom indicates a more spread-out distribution.
* **Standard Normal Distribution ((0, 1)):** The chi-square distribution arises from summing the squares of independent standard normal variables, each with a mean of 0 and a standard deviation of 1.

**Example 1: Squared Standard Normal Variable (One Degree of Freedom)**

Consider a standard normal variable X ~ (0, 1). Let Z = (X - ) ~ (0, 1), where  is the mean (here, 0). Then, the squared standard normal variable Z follows a chi-square distribution with one degree of freedom ():

Z ~ 

This signifies that the probability distribution of Z can be described by the chi-square distribution with 1 degree of freedom.

**Example 2: Sum of Squares of Multiple Normal Random Variables (More Degrees of Freedom)**

Suppose we have n independent standard normal variables, denoted as Z, Z, ..., Z. We calculate the sum of their squares:

S = Z + Z + ... + Z

Remarkably, this sum, S, follows a chi-square distribution with n degrees of freedom ():

S ~ 

This property holds true even if the normal variables have different means (  0) but the same standard deviation ( = 1).

**Key Properties of Chi-Square Distribution:**

* **Non-negative:** The chi-square distribution takes only non-negative values (  0), as squaring a real number always results in a non-negative value.
* **Skewed to the right:** The distribution is positively skewed, meaning it has a longer tail towards larger values.
* **Shape influenced by degrees of freedom:** Higher degrees of freedom lead to a more symmetrical and bell-shaped distribution, approaching a normal distribution for a sufficiently large number of degrees of freedom.

**Applications of Chi-Square Distribution:**

* **Goodness-of-fit tests:** Evaluating if observed data frequencies align with expected frequencies under a specific hypothesis.
* **Test of independence:** Determining if two categorical variables are statistically independent.
* **Test of variance:** Assessing if a population variance is equal to a hypothesized value.

## The Fisher's F-Distribution (F)

The Fisher's F-distribution (denoted by F), also known as the F-ratio distribution, plays a crucial role in hypothesis testing related to variances.

**Understanding the F-Distribution:**

* **Degrees of freedom (n, p):** The F-distribution has two degrees of freedom parameters: n for the numerator and p for the denominator.
* **Ratio of independent chi-square variables:** It is the ratio of two independent chi-square variables, scaled by their respective degrees of freedom.

**Example:**

Let X ~  and Y ~  be independent chi-square variables with n and p degrees of freedom, respectively. Then, the F-distribution arises as:

F = (X / n) / (Y / p) = (X * p) / (Y * n)

**Applications of F-Distribution:**

* **Test of equality of variances:** Determining if two population variances are statistically equal.
* **Analysis of variance (ANOVA):** Assessing if there are significant differences among means of multiple groups.
