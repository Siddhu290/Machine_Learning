## Handling Exception

Exception handling is a programming technique used to manage and respond to runtime errors or unexpected conditions in a controlled way. It allows a program to continue running or to fail gracefully instead of crashing. This is typically done using `try`, `except`, `else`, and `finally` blocks in many programming languages.

## Need For Exception Handling

Exception handling is necessary to manage errors gracefully, ensuring a program can handle unexpected situations without crashing and providing meaningful feedback to the user or system.

##Process of Handling Exception



1. **Try**: Attempt to execute a block of code that might raise an exception.
   
2. **Except**: Specify what to do if a specific exception occurs within the `try` block.
   
3. **Else**: (Optional): Execute code if no exceptions were raised in the `try` block.
   
4. **Finally**:(Optional): Execute cleanup code that runs regardless of whether an exception occurred.

```python
print("Practicing for try block")
try:
    numerator = 50
    denom = int(input("Enter the denominator: "))
    quotient = numerator / denom
    print(quotient)
    print("Division performed successfully")
except ZeroDivisionError:
    print("Denominator as ZERO.... not allowed")
print("OUTSIDE try..except block")


```

##Use of multiple except clauses

```python
 print ("Handling multiple exceptions")
 try:
    numerator=50
    denom=int(input("Enter the denominator: "))
    print (numerator/denom)
    print ("Division performed successfully")
 except ZeroDivisionError:
    print ("Denominator as ZERO is not allowed")
 except ValueError:
    print ("Only INTEGERS should be entered")
```

##Use of except without specifying an exception

```python
 print ("Handling exceptions without naming them")
 try:
    numerator=50
    denom=int(input("Enter the denominator"))
    quotient=(numerator/denom)
    print ("Division performed successfully")
 except ValueError:
    print ("Only INTEGERS should be entered")
 except:
    print(" OOPS.....SOME EXCEPTION RAISED")
```

##Use of else clause

```python
 print ("Handling exception using try...except...else")
 try:
    numerator=50
    denom=int(input("Enter the denominator: "))
    quotient=(numerator/denom)
    print ("Division performed successfully")
 except ZeroDivisionError:
    print ("Denominator as ZERO is not allowed")
 except ValueError:
    print ("Only INTEGERS should be entered")
 else:
   print ("The result of division operation is ", quotient)
```

 ## Finally clause



1. **Execution Guarantee**: Executes code regardless of whether an exception is raised or not.
   
2. **Cleanup Tasks**: Ideal for releasing resources like files or network connections.
   
3. **Optional**: Can be used independently or alongside `try-except` blocks for comprehensive error handling.

```python
print("Handling exception using try...except...else...finally")
try:
    numerator = 50
    denom = int(input("Enter the denominator: "))
    quotient = numerator / denom
    print("Division performed successfully")
except ZeroDivisionError:
    print("Denominator as ZERO is not allowed")
except ValueError:
    print("Only INTEGERS should be entered")
else:
    print("The result of division operation is ", quotient)
finally:
    print("OVER AND OUT")
```
