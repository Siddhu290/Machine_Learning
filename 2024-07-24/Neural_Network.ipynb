## Deep Learning

**Subset of Machine Learning:** Deep learning is a powerful technique within machine learning that uses artificial neural networks with multiple layers.

**Learning from Layers:**  These layers process information progressively, extracting increasingly complex features from the data.

**Tackling Complex Tasks:**  This allows deep learning to excel at tasks like:

* Image recognition (identifying objects in pictures)
* Natural language processing (understanding and generating human language)
* Speech recognition (converting spoken words to text)

**Revolutionizing AI:** Deep learning is a major driver of advancements in artificial intelligence.

---


**Neural Network**

**Inspired by the brain:** A neural network is a system of interconnected nodes that mimics the way biological neurons work together.

**Learning machine:** It processes information and learns from data, allowing it to improve at specific tasks over time.

**Examples:** Image recognition, speech recognition, making predictions.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
data = np.random.random((1000, 100))
label = np.random.randint(2, size=(1000, 1))


model = Sequential()
model.add(Dense(64,activation="relu",input_dim=100))
model.add(Dense(1,activation="sigmoid"))
model.compile(optimizer="rmsprop",loss="binary_crossentropy",metrics=["accuracy"])
model.fit(data,label,epochs=10,batch_size=32)
predictions = model.predict(data)
model.save("model.h5")
```
