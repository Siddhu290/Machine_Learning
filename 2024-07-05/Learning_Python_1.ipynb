**Byte Code:-** Python first compiles your source code into a form as byte code.

Python process has write access on your machine, it will store the byte code if your porgrams in files that end with a .pyc extension.

Compilation is simply a translation step, and byte code is lower-level,form-independent representation of your source code.

**PVM(Python Virtual Machine)**:- Once your program compiled to byte code,it is shipped for execution to something generally known as the PVM.

**CPython**:- The original and standard, implemeantation of python is usually call CPython,when you want to contrast it with the other two.Its name comes from the fact that it is coded in portable ANSI C language code.

**JPython**:- The jpython system is an alternative implementation of the python language ,targeted for integration with the Java Programming language.Jython consists of Java classes that compile Python source code to java code and then route the resulting byte code to the JAVA VIRTUAL MACHINE(JVM).

**IronPython**:- Is designed to allow Python programs to integrate with applications coded to work with Microsoft .NET Framework for Windows, as well as the Mono open source equivalent for Linux .NET and its C# programming language.

**Psyco:- Just in time compiler.**

**Frozen Binaries**:- it is possible to turn your Python programs into true executables ,known as Frozen Binaries in the python world.
