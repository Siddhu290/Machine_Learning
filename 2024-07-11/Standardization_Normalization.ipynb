##Chi-Square-Hypothesis

**1].Purpose:** Tests if there's a significant association between two categorical variables.

**2].Hypotheses:**
Null (H₀): No association between variables (independent).
Alternative (H₁): Association exists (dependent).

**3].Steps:**
Construct a contingency table with observed frequencies.

Calculate expected frequencies under H₀.

Compute the Chi-Square statistic
𝜒

 .

**4].Decision Rule:**
Determine degrees of freedom (df).
Compare the Chi-Square statistic to the critical value or p-value to reject or fail to reject H₀.

**5].Interpretation:**
If p-value ≤ significance level (e.g., 0.05), reject H₀ (association exists).
If p-value > significance level, fail to reject H₀ (no association).

```python
x=df.iloc[:,[3,4]].values
```

```python
y=df.iloc[:,3].values
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans()
```

```python
scale=StandardScaler()
```

```python
X_train=scale.fit_transform(X_train)
```

```python
y_train=scale.fit_transform(y_train.reshape(-1,1))
```

```python
model.fit(X_train,)
```

```python

import matplotlib.pyplot as plt
wcss=[]

for i in range (1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(X_train)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
y_kmeans=kmeans.fit_predict(x)

plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='Cluster1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='blue',label='Cluster2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='green',label='Cluster3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=100,c='cyan',label='Cluster4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=100,c='magenta',label='Cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='black',label='Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
