#**Model Building**
---
*Logistic Regression*

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode= LabelEncoder()
```

```python
df
```

```python


features = df[['HomePlanet', 'CryoSleep', 'Destination', 'Age', 'VIP',
            'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck',
            'Grouped', 'Deck', 'Side', 'Is_Embryo']]
```

```python
x=pd.get_dummies(features)
y=df[['Transported']]
```

```python
df['Transported']=encode.fit_transform(df['Transported'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model= LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_pred, y_test), annot=True, fmt="d", cmap="Blues", cbar=False)


plt.xlabel("Predicted Label", fontsize=14)
plt.ylabel("True Label", fontsize=14)
plt.title("Confusion Matrix for Logistic Regression Model", fontsize=16)


plt.show()
```
