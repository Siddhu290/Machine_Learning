#***Keras Sequential Model***


#When to use a Sequential Model:

    A Sequential model is appropriate for a plain stack of layers
    where each layer exactly one input layer tensor and one output Sensor.

A Sequential model is not appropriate when:

    Your model has multiple inputs or multiple outputs
    Any of your layers has multiple inputs or multiple outputs
    You need to do layer sharing
    You want non-linear topology (e.g. a residual connection, a multi-branch model) *italicized text*

````python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Activation
from tensorflow.keras.activations import relu
````

````python

# Assuming 'data' and 'one_hot_labels' are numpy arrays that you load or create
# For demonstration, let's create some dummy data
import numpy as np
data = np.random.rand(100, 784)  # 100 samples with 784 features
one_hot_labels = np.random.randint(2, size=(100, 10))  # 100 labels, one-hot encoded

model=Sequential()
model.add(Dense(32,input_dim=784))
model.add(Activation('relu'))
model.add(Dense(10))
model.add(Activation('softmax'))

model.compile(optimizer="rmsprop",loss='binary_crossentropy', metrics=['accuracy'])

# Do not overwrite the 'model' variable with the result of model.fit
history = model.fit(data,one_hot_labels,epochs=10,batch_size=32)

# Assuming 'X' is your test data, make sure it's defined and has the correct shape
X = np.random.rand(50, 784) # Example test data
Y=model.predict(X)


````
